```python
import requests
from bs4 import BeautifulSoup
import html5lib
import pandas as pd

url="https://www.mediamarkt.nl/nl/category/_laptops-482723.html?searchParams=&sort=&view=&page="
comp_name,price, description =[],[],[]

for page in range(1,21):
  html = requests.get(url + str(page)).content
  soup=BeautifulSoup(html,"html.parser")

  liste=soup.find_all("div",{"class":"product-wrapper"})

  for i in liste:
    price.append(i.find("div",{"class":"price small"}).text.strip("-,"))
    comp_name.append(i.h2.text.strip().split("-")[0])
    details_title =i.find("dl",{"class":"product-details"}).find_all("dt")
    details =i.find("dl",{"class":"product-details"}).find_all("dd")
    merge=[]
    for s1,s2 in zip(details_title,details):
      merge.append(s1.text + s2.text)
    description.append(''.join(merge))

keys=['comp_name',
      'price',
      'description']
values=[comp_name,
        price, 
        description]
dictionary = dict(zip(keys, values))

df = pd.DataFrame.from_dict(dictionary)
df
```
